# Strategy Bin High Î² = 1.0, k = 52

```python
# compute similarity matrix from ctx
ctx = Context.fromfile("/Users/monicasen/PycharmProjects/topic-modeling/datasets/rychkova_papers/matrix/rychkova_abstracts_lemmas_en_filtered__20250822-064803/bin_high_10/cxt")
sim_matrix = compute_conceptual_similarity_matrix(ctx)

# hca clustering
Z, _, _ = compute_hierarchical_clusters(sim_matrix)
```

```python
rows = []
k_min, k_max = 4, 80

for k in range(k_min, k_max + 1, 5):
    sil, labels_k = compute_silhouette_for_k(sim_matrix, Z, k=k)
    stats = cluster_stats(labels_k)
    rows.append({
        "k": k,
        "silhouette": sil,
        **stats
    })

sil_df = pd.DataFrame(rows)
sil_df
```

```python
plot_silhouette_curve(sil_df)
```

```python
# number of clusters
k = 52

# hca clustering
Z, clusters, labels = compute_hierarchical_clusters(sim_matrix, n_clusters=k)

plot_dendrogram(Z, labels, title="Dendrogram")
```

```python
pd.set_option('display.max_colwidth', None)

df_clusters = clusters_to_df(clusters)
display(df_clusters)
```
